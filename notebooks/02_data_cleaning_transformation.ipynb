```python
import pandas as pd

base_path = '/content/drive/MyDrive/healthcarefraud_detection/data/raw'

train = pd.read_csv(f'{base_path}/Train.csv')
test = pd.read_csv(f'{base_path}/Test.csv')
train_bene = pd.read_csv(f'{base_path}/Train_Beneficiarydata.csv')
test_bene = pd.read_csv(f'{base_path}/Test_Beneficiarydata.csv')
train_inp = pd.read_csv(f'{base_path}/Train_Inpatientdata.csv')
test_inp = pd.read_csv(f'{base_path}/Test_Inpatientdata.csv')
train_out = pd.read_csv(f'{base_path}/Train_Outpatientdata.csv')
test_out = pd.read_csv(f'{base_path}/Test_Outpatientdata.csv')

```

```python
from IPython.display import display
import pandas as pd

def column_sample_report_transposed(df, df_name, n=5):
    """
    Generates a transposed tabular report of columns, their types, and sample values.

    Args:
        df (pd.DataFrame): The DataFrame to report on.
        df_name (str): The name of the DataFrame for the report title.
        n (int): The number of unique values to sample per column.
    """
    print(f"\n=== {df_name} ===")

    report_data = {"Information Type": ["Column Name", "Data Type"] + [f"Sample {i+1}" for i in range(n)]}

    for col in df.columns:
        dtype = df[col].dtype
        vals = df[col].dropna().unique()
        sample = list(vals[:n]) + [None] * (n - len(vals[:n])) # Pad with None if less than n samples

        report_data[col] = [col, dtype] + sample

    report_df = pd.DataFrame(report_data)
    report_df = report_df.set_index("Information Type")
    display(report_df)

# Repeat for each DataFrame you care about using the new function
column_sample_report_transposed(train_inp, "Train_Inpatientdata")
column_sample_report_transposed(train_out, "Train_Outpatientdata")
column_sample_report_transposed(train_bene, "Train_Beneficiarydata")
column_sample_report_transposed(train, "Train")
```

```python
# Inpatient Claims
date_cols_inp = ['ClaimStartDt', 'ClaimEndDt', 'AdmissionDt', 'DischargeDt']
for col in date_cols_inp:
    train_inp[col] = pd.to_datetime(train_inp[col], errors='coerce')

# Outpatient Claims
date_cols_out = ['ClaimStartDt', 'ClaimEndDt']
for col in date_cols_out:
    train_out[col] = pd.to_datetime(train_out[col], errors='coerce')

# Beneficiary Data
date_cols_bene = ['DOB', 'DOD']
for col in date_cols_bene:
    train_bene[col] = pd.to_datetime(train_bene[col], errors='coerce')
```

```python
print(train_inp[date_cols_inp].dtypes)
print(train_out[date_cols_out].dtypes)
print(train_bene[date_cols_bene].dtypes)
```

```python
train['PotentialFraud'] = train['PotentialFraud'].map({'Yes': 1, 'No': 0})
print(train['PotentialFraud'].value_counts())
```

```python
train_bene['RenalDiseaseIndicator'] = train_bene['RenalDiseaseIndicator'].map({'Y': 1, '0': 0})
print(train_bene['RenalDiseaseIndicator'].value_counts(dropna=False))
```

```python
# Inpatient Claims
train_inp['DeductibleAmtPaid'] = train_inp['DeductibleAmtPaid'].fillna(0)
train_inp['DiagnosisGroupCode'] = train_inp['DiagnosisGroupCode'].fillna('-1')
train_inp['ClmAdmitDiagnosisCode'] = train_inp['ClmAdmitDiagnosisCode'].fillna('Unknown')
```

```python
# Outpatient Claims
if 'DeductibleAmtPaid' in train_out.columns:
    train_out['DeductibleAmtPaid'] = train_out['DeductibleAmtPaid'].fillna(0)
if 'ClmAdmitDiagnosisCode' in train_out.columns:
    train_out['ClmAdmitDiagnosisCode'] = train_out['ClmAdmitDiagnosisCode'].fillna('Unknown')
```

```python
print("DeductibleAmtPaid nulls (inpatient):", train_inp['DeductibleAmtPaid'].isnull().sum())
print("DiagnosisGroupCode nulls (inpatient):", train_inp['DiagnosisGroupCode'].isnull().sum())
print("ClmAdmitDiagnosisCode nulls (inpatient):", train_inp['ClmAdmitDiagnosisCode'].isnull().sum())

print("DeductibleAmtPaid nulls (outpatient):", train_out['DeductibleAmtPaid'].isnull().sum())
print("ClmAdmitDiagnosisCode nulls (outpatient):", train_out['ClmAdmitDiagnosisCode'].isnull().sum())
```

```python
from IPython.display import display
import pandas as pd

def transposed_schema_report(df, df_name):
    """
    Generates a transposed tabular report of column names and their types.

    Args:
        df (pd.DataFrame): The DataFrame to report on.
        df_name (str): The name of the DataFrame for the report title.
    """
    print(f"\n--- {df_name} Schema (Transposed) ---")

    # Get the dtypes as a Series
    dtypes_series = df.dtypes

    # Convert the Series to a DataFrame
    # The index (column names) becomes a column, and the values (dtypes) become a column
    schema_df = dtypes_series.reset_index()
    schema_df.columns = ['Column Name', 'Data Type']

    # Set 'Column Name' as the index
    schema_df = schema_df.set_index('Column Name')

    # Transpose the DataFrame
    transposed_df = schema_df.transpose() # The .T accessor also works here

    display(transposed_df)

# Apply the transposed schema report to each DataFrame
transposed_schema_report(train_inp, "Train_Inpatientdata")
transposed_schema_report(train_out, "Train_Outpatientdata")
transposed_schema_report(train_bene, "Train_Beneficiarydata")
transposed_schema_report(train, "Train")
```
